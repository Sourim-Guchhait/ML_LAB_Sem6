```python
# importing libraries  
import numpy as nm
import matplotlib.pyplot as mtp
import pandas as pd


#importing datasets  
data_set=pd.read_csv("C:/Users/SOURIM GUCHHAIT/my_data.csv")
data_set
```

```python
#extracting indepndent and dependent variable
x=data_set.iloc[:,[2,3]].values
y=data_set.iloc[:,4].values
```

```python
#spliting the data_set into training and test set.
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)

```

```python
x_train
```

```python
x_test
```

```python
data_set
```

```python
#featuring scaling 
from sklearn.preprocessing import StandardScaler
st_x=StandardScaler()
x_train=st_x.fit_transform(x_train)
x_test=st_x.fit_transform(x_test)
```

```python
#fitting teh random forest algorithm to the trainning set
#fitting decitsion tree classifier to the training set
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier(n_estimators=10,criterion="entropy")
classifier.fit(x_train,y_train)

                                  
```

```python
#predicting the set result
y_pred=classifier.predict(x_test)
```

```python
y_pred
```

```python
#creating the confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
#visualizing the trainning set result
from matplotlib.colors import ListedColormap
x_set,y_set=x_train,y_train
x1,x2=nm.meshgrid(nm.arange(start = x_set[:,0].min()-1,stop=x_set[:,0].max()+1,step =0.01),
nm.arange(start=x_set[:,1].min()-1,stop = x_set[:,1].max() +1,step = 0.01))

mtp.contourf(x1,x2,classifier.predict(nm.array([x1.ravel(),x2.ravel()]).T).reshape(x1.shape),
alpha=0.75,cmap = ListedColormap(('yellow','green')))
mtp.xlim(x1.min(),x1.max())
mtp.xlim(x2.min(),x2.max())
for i,j in enumerate(nm.unique(y_set)):
    mtp.scatter(x_set[y_set==j,0],x_set[y_set==j,1],
                
            
c=ListedColormap(('purple','green'))(i),label=j)
mtp.title('Random Forest Algorithm(Tranning set)')
mtp.legend()
mtp.show()
```
